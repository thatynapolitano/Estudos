#### Como encontrar o Big-O?

- Como efetivamente encontrandos os Big-O dos algoritmos?

- Analisando os algoritmos com base nos dados de entrada (n)



#### Algoritmos com Laço Simples
- O(n)
- São algoritmos que podem executar "n" vezes um determinado código:

```python
def ImprimeLista(lista):
    for i in lista:#0(n) 
        print(i)#0(1) 

print(ImprimeLista([4,5,6,7]))#0(1) 
```

#### Propriedade da Adição
- O(n)




```python
def ImprimeLista(lista1,lista2):
    for i in lista1:#0(n) 
        print(i)#0(1) 
    for i in lista2:#0(n) 
        print(i)#0(1) 

print(ImprimeLista([2,4,6],[8,9,10]))#0(1) 
```

#### Propriedade da Multiplicação
- O(n²)

```python
def SomaElementos(lista1, lista2):  # PA - progracao aritmetica
    for i in lista1:#0(n^2) 
        for j in lista2:#0(n) 
            print(i+j)#0(1) 

print(SomaElementos([2,4,6],[1,2,2]))
```

#### Progressão Aritmética (PA)

- Sequência de números na qual a diferença entre os dois termos consecutivos é constante.

1+3+5=7=9+11+...+n <br>
9+6+3+0-3-6...-n <br>
1+2+3+4+5+6...+n <br>



```python
# Progressão Geométrica (PG)

def boo(lista1):
    for i in lista1: #0(n) 
        for j in range(0, i*2, 1): #0(n) 
            print(i+j)  #0(1) 

print(boo([1,2,3]))#0(1) 
```

#### Busca Binária Algoritmo

```python
def buscaBinaria(dados, buscado):
    inicio = 0
    fim = len(dados)

    while inicio <= fim:
        meio = int((inicio+fim)/2)
        if buscado > dados[meio]:
            inicio = meio + 1
        elif buscado < dados [meio]:
            fim = meio -1
        else: 
            return meio
    return -1 


print(buscaBinaria([2,3,4,5,6],4))
```
