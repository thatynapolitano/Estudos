### Bubble Sort - Algoritmo de ordenação 

O objetivo desse algoritmo é ordenar os dados. (Crescente ou decrescente)
Também conhecido por:
- Ordenação bolha
- Método bolha
- Ordenação por troca

Independemtente da linguagem de programação, a lógica do algoritmo é a mesma!
Isso vale para outros algoritmos de ordenação!

#### Como funciona o Bubble Sort?

- O algoritmo Bubble Sort varre o conjunto de dados do início ao fim, comparando os dados de dois em dois e efetuando a troca, se necessário.
- A varredura ocorrerá uma quantidade de vezes igual ao tamanho do conjunto de dados (n). 




Ocorrem comparações entre os números de um vetor.

A quantidade de varredura é igual ao tamanho do vetor. Então no exemplo acima, terão 5 varreduras.

```python
dados = [6, 7, 5, 1, 2]

def bubbleSort(dados):
    n = len(dados)
    for vez in range(n):
        for i in range(n-1):
            print(f"V ={vez}: Comparando {dados[i]} com {dados[i+1]}", end="")
            if dados[i] > dados[i+1]:
                print("Troca? SIM")
                aux = dados[i]
                dados[i] = dados[i+1]
                dados[i+1] = aux 
            else:
                print("Troca? NÃO")
 
print(f"Entrada: {dados}")
bubbleSort(dados)
print(f"Saída: {dados}")
```
