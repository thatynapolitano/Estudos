### Bubble Sort - Algoritmo de ordenação 

O objetivo desse algoritmo é ordenar os dados. (Crescente ou decrescente)
Também conhecido por:
- Ordenação bolha
- Método bolha
- Ordenação por troca

Independemtente da linguagem de programação, a lógica do algoritmo é a mesma!
Isso vale para outros algoritmos de ordenação!

#### Como funciona o Bubble Sort?

- O algoritmo Bubble Sort varre o conjunto de dados do início ao fim, comparando os dados de dois em dois e efetuando a troca, se necessário.
- A varredura ocorrerá uma quantidade de vezes igual ao tamanho do conjunto de dados (n). 




Ocorrem comparações entre os números de um vetor.

A quantidade de varredura é igual ao tamanho do vetor. Então no exemplo acima, terão 5 varreduras.

```python
dados = [6, 7, 5, 1, 2]

def bubbleSort(dados):
    n = len(dados)
    for vez in range(n):
        for i in range(n-1):
            print(f"V ={vez}: Comparando {dados[i]} com {dados[i+1]}", end="")
            if dados[i] > dados[i+1]:
                print("Troca? SIM")
                aux = dados[i]
                dados[i] = dados[i+1]
                dados[i+1] = aux 
            else:
                print("Troca? NÃO")
 
print(f"Entrada: {dados}")
bubbleSort(dados)
print(f"Saída: {dados}")
```

### Merge-Sort - Dividir para conquistar 

- O Merge-sort utiliza do princípio de Dividir para conquistar!
- Ele divide o conjunto de dados sempre ao meio, até que não seja mais possível dividir (caso base). 

meio = int(inicio+fim/2)

### Chamadas Recursivas 

- Total de conjuntos por nivel:



- Total de chamadas recursivas da função Merge Sort para um conjunto inicial de tamanho 8:


```python
def mergeSort(dados):
    if len(dados)>1:
        meio = len(dados)//2
        esquerda = dados[meio:]
        direita = dados[:meio]
        mergeSort(esquerda)
        mergeSort(direita)
        i = j = k = 0
        while i < len(esquerda) and j < len(direita):
            if esquerda[i] < direita[j]:
                dados[k] = esquerda[i]
                i+=1 
            else:
                dados[k] = direita [j]
                j+=1
            k+=1
        while i < len(esquerda):
            dados[k] = esquerda [i]
            i+=1
            k+=1
        while j<len(direita):
            dados[k]=direita[j]
            j+=1
            k+=1

dados = [54, 26, 93, 17, 77, 31, 44, 55]
mergeSort(dados)
print(dados)
```
