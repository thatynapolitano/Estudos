### Plotando gráficos

#### .plot()

A biblioteca pandas tem a função [pd.DataFrame.plot()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html) para traçar gráficos de Series ou DataFrame. Usamos um datafame com valores de [qualidade do ar](https://github.com/pandas-dev/pandas/blob/main/doc/data/air_quality_no2.csv) em três estações de trem da europa

```python
csv_file = "../Data/air_quality_no2.csv"

air_quality = pd.read_csv(
    filepath_or_buffer = csv_file,
    index_col = 0,
    parse_dates = True
)

display(air_quality.head(500))

air_quality.plot()
plt.show();
```

Podemos plotar uma estação específica.

```python
air_quality['station_paris'].plot()
plt.axhline(y = air_quality['station_paris'].median(), color = "red", linestyle = "--")
plt.axhline(y = air_quality['station_paris'].mean(), color = "blue", linestyle = "--")
plt.show()
```

```python
fig, axes = plt.subplots(nrows = len(air_quality.columns), ncols = 1, figsize = (4, 12))

# Plotting data and adding median lines
for i, col in enumerate(air_quality.columns):
    
    air_quality[col].plot(ax = axes[i], kind = 'area')
    median = air_quality[col].median()
    mean = air_quality[col].mean()
    axes[i].axhline(y = median, color = 'r', linestyle = '--')
    axes[i].axhline(y = mean, color = 'g', linestyle = '--')
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

### Criando um novo dataframe

Vamos criar um novo dataframe a partir da rotina [pd.DataFrame()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html)

```python
df = pd.DataFrame(
    data = {
        'A' : {0 : 'a', 1 : 'b', 2: 'c'},
        'B' : {0 : 1, 1 : 3, 2 : 5},
        'C' : {0 : 2, 1 : 4, 2 : 6}
    }
)

df
```

Note os parâmetros em [pd.melt()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.melt.htmlNote), 


- `id_vars` : Coluna(s) a serem usadas como variáveis identificadoras, 
- `value_vars` : Coluna(s) para realizar a remoção do pivoteamento. Se não for especificado, usa todas as colunas que não estão definidas como `id_vars`.

```python
pd.melt(frame = df, id_vars = ["A"], value_vars = ["B", "C"])
```

### Air Quality

Vamos ler o arquivo '../Data/air_quality_no2_long.csv' e plotar adistribuição diária de dióxido de nitrogênio na atmosfera de Paris.

```python
air_quality_2 = pd.read_csv(
    filepath_or_buffer = '../Data/air_quality_no2_long.csv',
    encoding = 'utf-8',
)
air_quality_2
```

```python
air_quality_2 = air_quality_2.rename(
    columns = {'date.utc' : 'datetime'},
)
air_quality_2
```

```python
air_quality_2['datetime'] = pd.to_datetime(air_quality_2['datetime'])
air_quality_2
```

```python
air_quality_2['location'].unique()
```

```python
fig, axs = plt.subplots(figsize = (12, 4))

city_test = 'Paris'
air_quality_city_test = air_quality_2[air_quality_2['city'] == city_test].groupby([air_quality_2['datetime'].dt.hour])['value'].mean()
#air_quality_city_test

air_quality_city_test.plot(
    kind = 'bar',
    rot = 0,
    ax = axs
)

plt.title(f'{city_test}')
plt.xlabel('Hora do dia')
plt.ylabel('$NO_{2} (\mu/m^{3})$')
```

```python
fig, axs = plt.subplots(figsize = (12, 4))

city_test = 'London'
air_quality_city_test = air_quality_2[air_quality_2['city'] == city_test].groupby([air_quality_2['datetime'].dt.hour])['value'].mean()
#air_quality_city_test

air_quality_city_test.plot(
    kind = 'bar',
    rot = 0,
    ax = axs
)

plt.title(f'{city_test}')
plt.xlabel('Hora do dia')
plt.ylabel('$NO_{2} (\mu/m^{3})$')
```
